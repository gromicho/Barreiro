```python
from pathlib import Path
import toml

def validate_with_streamlit_parser(path: str) -> None:
    """Validate a secrets.toml file using the same 'toml' parser Streamlit uses."""
    p = Path(path)
    text = p.read_text(encoding='utf-8', errors='replace')
    try:
        toml.loads(text)
        print('OK: toml parser (Streamlit) accepts this file.')
    except Exception as exc:
        print(f'FAIL: {exc}')
        print('--- Context around reported location (approx) ---')
        # If the exception message has "line X", print a small window around it:
        lines = text.splitlines()
        for i in range(48, min(len(lines), 60)):
            print(f'{i+1:>4}: {lines[i]}')


validate_with_streamlit_parser(
    r'C:\Users\joaqu\OneDrive - UvA\Documenten\GitHub\anekke\Barreiro\.streamlit\secrets.toml'
)
```
